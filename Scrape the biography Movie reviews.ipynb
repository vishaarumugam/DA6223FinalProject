```python
# Get the page content in a HTML Format
def get_webpage_content(url):
    response = requests.get(url)
    page_content = BeautifulSoup(response.text,'html.parser')
    return (response, page_content)
```

```python
# Get the Ajax Pagination Next Page URL
def check_Load_more(web_content):
    next_page_url = ""
    if web_content.find('div',{'class':'load-more-data'}):
        data_ajaxurl = web_content.find('div',{'class':'load-more-data'})['data-ajaxurl']
        data_key = " "
        try:
            data_key = web_content.find('div',{'class':'load-more-data'})['data-key']
            next_page_url = base_url+data_ajaxurl+review_concat+data_key
        except KeyError:
            next_page_url = " "
            print("end of the page")
        return next_page_url
    else:
        next_page_url =""
        return next_page_url
```

```python
# Get the Review details 
def get_review_details(movie_name_input,Rev_content):
    Mov_name = movie_name_input
    Review_Det = Rev_content.findAll('div',{'lister-item-content'})
    for i in Review_Det:
        Rev_title= i.find('a').text.strip()
        Rev_title = Rev_title.translate(str.maketrans('','',string.punctuation))
        Rev_user = i.find('span',{'display-name-link','a'}).text
        if i. find('div',class_='text show-more__control'):
            Review_content = i. find('div',class_='text show-more__control').text
            Review_content = Review_content.translate(str.maketrans('','',string.punctuation)) 
            Review_content = Review_content.replace("\n","")
        else:
            Review_content = "Not Applicable"
        if i.find('div',{'ipl-ratings-bar'}):
            Rating = i.find('span').text.strip().split('/')[0]
        else:
            Rating = "NA"
        
        myfile1 =open("Review.csv", 'a', newline="",encoding='utf8', errors = "ignore")
        csvwriter = csv.writer(myfile1)
        myfile1.write(Mov_name+','+Rev_title+','+Rev_user+','+Review_content+','+Rating+','+'\n')
        myfile1.close()
```

# Annotate the Dataset with the Gold Label and Sentiment Score

```python
def map_gold_label(rating):
    if rating == 5:
        return 'Neutral'
    elif rating < 5:
        return 'Negative'
    else:
        return 'Positive'

def map_Sentiment(label):
    if label == 'Positive':
        return 1
    elif label == 'Negative':
        return -1
    else:
        return 0
```

```python
All_movie_review['Gold_Label']=[ map_gold_label(x) for x in All_movie_review['Rating']]
All_movie_review['Sentiment']=[ map_Sentiment(x) for x in All_movie_review['Gold_Label']]
All_movie_review.head()
```

```python
# Remove the punctuation if any and get the clean Tokens
import nltk
from nltk.corpus import stopwords
import string

def remove_punct(text):
    no_punct = [c for c in text if c not in string.punctuation]
    return no_punct

All_movie_review['Clean_Tokens'] = [remove_punct(w) for w in All_movie_review["lemmatized_tokens"]] 
print(All_movie_review['Clean_Tokens'])
```
